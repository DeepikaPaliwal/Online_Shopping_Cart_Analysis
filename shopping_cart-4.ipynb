```python
#Univariant data analysis
def uni_data(data,color,title1,title2):       #defining function
    fig, (ax1, ax2) = plt.subplots( # subplots
        ncols=2, # num of cols
        nrows=1, # num of rows
        figsize=(20, 6) # set the width and high
    )
        ##below are two plots: histplot and qqplot
    sns.histplot(                #ploting displot visuvlization
            data,
            color=color,
            kde=True,
            ax=ax1
    )

    ax1.set_title(
        title1,
        weight='bold',
        fontsize=15,
        pad=30
    )
    sm.qqplot(
        data,
        ax=ax2,
        line='s'
    )

    ax2.set_title(
        title2,
        weight='bold',
        fontsize=15,
        pad=30
    )
    return fig
```

```python
uni_data(
    cop_data['sales'],
    color='blue',
    title1= 'cop data-sales distribution',
    title2='Quantile plot'
);
```
